```python
import os
```

```python
path = os.path.join("Q1" , "Q1")
os.makedirs(path)
```

```python
for img in os.listdir("Q1"):
  if img.endswith('.jpeg'):
    #path of picture
    p = os.path.join("Q1" , img)
    hold = img.split('.')
    p2 = os.path.join("Q1/Q1", hold[0])
    
    #read the image then change color to gray and write it in the file 
    my_image = cv2.imread(p)
    gray_image = cv2.cvtColor(my_image, cv2.COLOR_BGR2GRAY)
    cv2_imshow(gray_image)
    cv2.imwrite(p2 + ".jpeg", gray_image)
    #store the shape of each picture
    f = open(p2 + ".txt" , "w")
    f.write(str(gray_image.shape))
  
```
